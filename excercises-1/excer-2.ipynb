# Thyroid Disease 1.2
-  Download the file allbp.data from the Thyroid Disease data set available at https://archive.ics.uci.edu/
dataset/102/thyroid+disease.

- 1.2 -> Clean the data and count characteristics

  

```python
import csv as C
import numpy as N
import pandas as P

# Strip lines 11-39 from allbp.names and assign them to column_names
with open('./thyroid+disease/allbp.names', 'r') as file:
    lines = file.readlines()
    column_names = [line.strip() for line in lines[10:39]]  
    column_names = [name.split(':')[0].strip() for name in column_names]

df = P.read_csv('./thyroid+disease/allbp.data',
                sep=",",
                na_values=["?"],
                index_col=False,
                header=None,
                names=column_names)

df
```

```python
column_names
```

2800 rows or datapoints -> 2800 observations
29 columns of data -> 29 variables

```python
# Convert 't' and 'f' to boolean
bool_columns = df.columns[df.isin(['t', 'f']).any()]
df[bool_columns] = df[bool_columns].replace({'t': True, 'f': False}).astype(bool)

df.dtypes
```

```python
for item in df.columns:
    print(item)
```

```python
# Make a dataframe where variable and its null count are displayed
count_nan = df.isnull().sum()
count_nan
```

```python
# Count variables with missing values

count_varibles_with_nan = len(count_nan[count_nan > 0])
count_varibles_with_nan
```

```python
# Counting sums of True for each variable that is boolean

df_booleans = df.select_dtypes(bool).sum()
print("Boolean attributes: " + str(len(df_booleans))), df_booleans




```

```python
# TSH T3 TT4 T4U FTI and TBG -> calculate sum of squared values divided by non-NaN values

columns = ['TSH', 'T3', 'TT4', 'T4U', 'FTI', 'TBG']
squared_values = df[columns] ** 2
squared_values  # Display squared values
# Sum all squared values
total_sum_squared = squared_values.sum(skipna=True)
total_sum_squared


sum_squared_ratio = squared_values.sum() / df[columns].count()
sum_squared_ratio



```
